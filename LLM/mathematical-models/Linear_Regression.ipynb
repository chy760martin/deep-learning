# 선형회귀 총정리

## 모델 정의
- 모델 식: $ \hat{y}_i = X_i W + b $

---

## 손실 함수 (MSE)
- 손실 함수: $ L(W, b) = \frac{1}{N} \sum_{i=1}^N (y_i - (X_i W + b))^2 $

---

## Gradient 계산 (체인룰 적용)

### W에 대한 미분 과정
1. 손실 함수 항: $ (y_i - (X_i W + b))^2 $
2. 제곱을 미분: $ 2(y_i - (X_i W + b)) \cdot \frac{\partial}{\partial W}(y_i - (X_i W + b)) $
3. 안쪽식 미분: $ \frac{\partial}{\partial W}(y_i - (X_i W + b)) = -X_i $
4. 결합: $ -2X_i(y_i - \hat{y}_i) $
5. 전체 공식: $ \frac{\partial L}{\partial W} = -\frac{2}{N} \sum_{i=1}^N X_i (y_i - \hat{y}_i) $

---

### b에 대한 미분 과정
1. 손실 함수 항: $ (y_i - (X_i W + b))^2 $
2. 제곱을 미분: $ 2(y_i - (X_i W + b)) \cdot \frac{\partial}{\partial b}(y_i - (X_i W + b)) $
3. 안쪽식 미분: $ \frac{\partial}{\partial b}(y_i - (X_i W + b)) = -1 $
4. 결합: $ -2(y_i - \hat{y}_i) $
5. 전체 공식: $ \frac{\partial L}{\partial b} = -\frac{2}{N} \sum_{i=1}^N (y_i - \hat{y}_i) $

---

## 예시 데이터
- 샘플1: $ X_1 = 1, y_1 = 5 $  
- 샘플2: $ X_2 = 2, y_2 = 8 $  
- 초기값: $ W = 2, b = 1 $

---

### (1) 순전파
- 예측값: $ \hat{y}_1 = 3, \hat{y}_2 = 5 $

---

### (2) 손실 함수
- 손실 값: $ L = \frac{1}{2}[(5-3)^2 + (8-5)^2] = 6.5 $

---

### (3) 역전파 (편미분 계산)
- W에 대한 Gradient: $ \frac{\partial L}{\partial W} = -\frac{2}{2}[1(5-3) + 2(8-5)] = -8 $  
- b에 대한 Gradient: $ \frac{\partial L}{\partial b} = -\frac{2}{2}[(5-3) + (8-5)] = -5 $

---

### (4) 가중치 업데이트 ($ \eta = 0.1 $)
- 업데이트 후: $ W = 2 - 0.1 \cdot (-8) = 2.8 $  
- 업데이트 후: $ b = 1 - 0.1 \cdot (-5) = 1.5 $
