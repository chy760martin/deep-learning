# MLP(Multi-Layer Perceptron, 다층 퍼셉트론 분류 모델) 3층 분류 모델 총정리 

## 1. 전체 흐름 요약
### 1) 순전파: 입력 → 은닉층1 선형결합 → ReLU → 은닉층2 선형결합 → ReLU → 출력층 선형결합 → Softmax → 손실 계산(Cross-Entropy)
### 2) 역전파: 손실 → 출력층 오차 → 출력층 기울기 → 은닉층 오차 → 은닉층 기울기
---
## 2. MLP 3층 Classification 특징
### 1) 입력층(Input Layer): 데이터 벡터 [ x ]
### 2) 은닉층1(Hidden Layer): 선형결합 + 비선형 활성화(ReLU 등)
### 3) 은닉층2(Hidden Layer): 선형결합 + 비선형 활성화(ReLU 등)
### 4) 출력층(Output Layer): 선형결합 + Softmax → 확률 분포
### 5) 손실 함수(Loss Function): Cross-Entropy → 정답 클래스 확률을 최대화하도록 학습
---
## 2. 순전파 공식
### 1) 은닉층1 선형결합 : 
$ [ z_1 = x W_1 + b_1 ] $
### 2) 은닉층1 활성화 (예: ReLU) : 
$ [ h1 = f(z_1) = \text{ReLU}(z_1) ] $
### 3) 은닉층2 선형결합 : 
$ [ z_2 = h1 W_2 + b_2 ] $
### 4) 은닉층 활성화 (예: ReLU) : 
$ [ h2 = f(z_2) = \text{ReLU}(z_2) ] $
### 5) 출력층 선형결합 : 
$ [ z_3 = h2 W_3 + b_3 ] $
### 6) 소프트맥스 변환 : 
$ [ y_k = \frac{e^{z_{2,k}}}{\sum_j e^{z_{2,j}}} ] $
### 7) 손실 함수 (Cross-Entropy) : 
$ 정답 클래스 [ c ]일 때 : [ L = -\log(y_c) ] $

---
## 3. 역전파 공식
### 1) 출력층 오차 계산 : 
$ [ \delta_3 = y - y_{\text{true}} ] $
### 2) 출력층 기울기 계산 : 
$ [ \frac{\partial L}{\partial W_3} = h_2^\top \delta_3 ] $
### 3) 은닉층2 오차 계산 :  
$ [ \delta_2 = \delta_3 W_3^\top \odot f^\prime(z_2) ] $
### 4) 은닉층2 기울기 계산 : 
$ [ \frac{\partial L}{\partial W_2} = h_1^\top \delta_2 ] $
### 5) 은닉층1 오차 계산 :  
$[ \delta_1 = \delta_2 W_2^\top \odot f^\prime(z_1) ] $
### 6) 은닉층1 기울기 계산 : 
$ [ \frac{\partial L}{\partial W_1} = x^\top \delta_1 ] $

---
## 4. 경사하강법(Gradient Descent) 공식(가중치 업데이트)
$ [ \eta ] $ : 학습률(Learning Rate), $ [ \frac{\partial L}{\partial W} ] $ : 손실 함수에 대한 가중치의 기울기(gradient)

손실 [ L ]을 최소화하기 위해 가중치 [ W ]를 업데이트하는 규칙은 다음과 같습니다 : 
$ [ W^{(new)} = W^{(old)} - \eta \cdot \frac{\partial L}{\partial W} ] $

바이어스도 동일하게 업데이트합니다: 
$ [ b^{(new)} = b^{(old)} - \eta \cdot \frac{\partial L}{\partial b} ] $

---
## 5. 핵심 요약 
	◦ MLP Classification 모델 목적 : MLP Classification은 입력을 받아 은닉층을 거쳐 Softmax로 확률 분포를 만들고, Cross-Entropy 손실을 통해 정답 클래스 확률을 최대화하도록 학습합니다.
	◦ 역전파는 출력층 오차에서 시작해 은닉층으로 전달되며, 각 층의 가중치 기울기를 계산해 업데이트합니다.
	◦ ReLU 미분: $ [ f^\prime(z) = 1 \; (z>0), \; 0 \; (z \leq 0) ] $
	◦ 출력층 오차는 소프트맥스 결과와 정답 벡터 차이
	◦ 기울기 계산은 항상 “이전 층 출력”과 “현재 층 오차”의 외적
	◦ 경사하강법으로 모든 가중치와 바이어스를 업데이트

## MLP 3층 분류 모델의 순전파 → 손실 계산 → 역전파 → 경사하강법 업데이트 과정을 공식과 함께 단계별로 계산

## 1. 순전파 (Forward Propagation)
---
### 1) 은닉층 1
$ [ z_1 = [1,2] \cdot \begin{bmatrix}0.1 & 0.2 \\ 0.3 & 0.4\end{bmatrix} = [0.7, 1.0] ] $

$ [ h_1 = \text{ReLU}(z_1) = [0.7, 1.0] ] $

---
### 2) 은닉층 2
$ [ z_2 = [0.7,1.0] \cdot \begin{bmatrix}0.5 & 0.6 \\ 0.7 & 0.8\end{bmatrix} = [1.05, 1.22] ] $

$ [ h_2 = \text{ReLU}(z_2) = [1.05, 1.22] ] $

---
### 3) 출력층
$ [ z_3 = [1.05,1.22] \cdot \begin{bmatrix}0.2 & 0.3 \\ 0.4 & 0.5\end{bmatrix} = [0.698, 0.925] ] $

---
### 4) 소프트맥스
$ [ y_1 = \frac{e^{0.698}}{e^{0.698}+e^{0.925}}, \quad y_2 = \frac{e^{0.925}}{e^{0.698}+e^{0.925}} ] $

$ [ e^{0.698} \approx 2.010, \quad e^{0.925} \approx 2.522 ] $

$ [ y = [0.444, 0.556] ] $

---
### 5) 손실 함수 (정답 클래스 = 2)
$ [ L = -\log(y_2) = -\log(0.556) \approx 0.587 ] $

---
## 2. 역전파 (Backpropagation)

---
### 1) 출력층 오차 계산
$ [ \delta_3 = y - y_{\text{true}} = [0.444, 0.556] - [0,1] = [0.444, -0.444] ] $

---
### 2) 출력층 기울기 계산
$ [ \frac{\partial L}{\partial W_3} = h_2^\top \cdot \delta_3 $

$ = \begin{bmatrix}1.05 \\ 1.22\end{bmatrix} \cdot [0.444, -0.444] ] $

$ [ = \begin{bmatrix}0.466 & -0.466 \\ 0.542 & -0.542\end{bmatrix} ] $

---
### 3) 은닉층 2 오차 계산
$ [ \delta_2 = \delta_3 W_3^\top \cdot f^\prime(z_2) ] $

$ [ = [0.444, -0.444] \cdot \begin{bmatrix}0.2 & 0.4 \\ 0.3 & 0.5\end{bmatrix} ] $

$ [ = [0.444\cdot0.2 + (-0.444)\cdot0.3, \; 0.444\cdot0.4 + (-0.444)\cdot0.5] ] $

$ [ = [0.0888 - 0.1332, \; 0.1776 - 0.222] = [-0.0444, -0.0444] ] $

---
### 4) 은닉층 2 기울기 계산
$ [ \frac{\partial L}{\partial W_2} = h_1^\top \cdot \delta_2 $

$ = \begin{bmatrix}0.7 \\ 1.0\end{bmatrix} \cdot [-0.0444, -0.0444] ] $

$ [ = \begin{bmatrix}-0.0311 & -0.0311 \\ -0.0444 & -0.0444\end{bmatrix} ] $

---
### 5) 은닉층 1 오차 계산
$ [ \delta_1 = \delta_2 W_2^\top \cdot f^\prime(z_1) ] $

$ [ = [-0.0444, -0.0444] \cdot \begin{bmatrix}0.5 & 0.7 \\ 0.6 & 0.8\end{bmatrix} ] $

$ [ = [-0.0222 - 0.0266, \; -0.0311 - 0.0355] = [-0.0488, -0.0666] ] $

---
### 6) 은닉층 1 기울기 계산
$ [ \frac{\partial L}{\partial W_1} = x^\top \cdot \delta_1 $

$ = \begin{bmatrix}1 \\ 2\end{bmatrix} \cdot [-0.0488, -0.0666] ] $

$ [ = \begin{bmatrix}-0.0488 & -0.0666 \\ -0.0976 & -0.1332\end{bmatrix} ] $

---
## 3. 경사하강법 (Gradient Descent)
학습률 $ [ \eta = 0.1 ] $

---

$ [ W_3^{new} = W_3^{old} - 0.1 \cdot \frac{\partial L}{\partial W_3} ] $

## 4. 최종 요약
	◦ 순전파: 입력 → 은닉층1 → 은닉층2 → 출력층 → Softmax → 손실
	◦ 역전파: 출력층 오차 → 은닉층2 오차 → 은닉층1 오차 → 각 층 기울기 계산
