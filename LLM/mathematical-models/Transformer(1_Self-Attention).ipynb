# Transformer(1.Self-Attention) м–ём–ҙ лӘЁлҚё мҙқм •лҰ¬ 

## 1. Self-Attention кіөмӢқ
---
### 1) Query, Key, Value кі„мӮ°(мғқм„ұ) : 
### $ [ Q = XW_Q, \quad K = XW_K, \quad V = XW_V  ] $
### м„ӨлӘ…)
мһ…л ҘлІЎн„° $ [ X = \begin{bmatrix}1 & 0 \\ 0 & 1\end{bmatrix} ] $

к°ҖмӨ‘м№ҳ н–үл ¬ кіұмңјлЎң Q, K, V мғқм„ұ :

Query: $ [ Q = XW_Q ] $

Key: $ [ K = XW_K ] $

Value: $ [ V = XW_V ] $

м—¬кё°м„ң $ [ W_Q, W_K, W_V ] $ лҠ” н•ҷмҠөлҗҳлҠ” нҢҢлқјлҜён„°мһ…лӢҲлӢӨ.

	в—Ұ QueryлҠ” вҖңлӮҙк°Җ лӢӨлҘё нҶ нҒ°м—җкІҢ л¬ҙм—Үмқ„ л¬јм–ҙліҙкі  мӢ¶мқҖк°Җ?вҖқ

	в—Ұ KeyлҠ” вҖңлӮҙк°Җ к°Җм§„ м •ліҙлҠ” л¬ҙм—Үмқёк°Җ?вҖқ

	в—Ұ ValueлҠ” вҖңлӮҙк°Җ мӢӨм ңлЎң м „лӢ¬н•  м •ліҙлҠ” л¬ҙм—Үмқёк°Җ?вҖқ
---
### 2) мң мӮ¬лҸ„(Score) кі„мӮ° : 
### $ [ \text{Score} = QK^\top ] $
### м„ӨлӘ…)
QueryмҷҖ KeyлҘј лӮҙм Ғ(dot product)н•ҙм„ң мң мӮ¬лҸ„лҘј кө¬н•©лӢҲлӢӨ:
$ [ \text{Score} = QK^\top ] $

	в—Ұ QueryлҠ” вҖңм§Ҳл¬ёмһҗвҖқ

	в—Ұ KeyлҠ” вҖңлӢөліҖмһҗвҖқ

	в—Ұ лӮҙм Ғ кІ°кіјлҠ” вҖңм§Ҳл¬ёмһҗмҷҖ лӢөліҖмһҗк°Җ м–јл§ҲлӮҳ мһҳ л§һлҠ”к°Җ?вҖқлҘј мҲҳм№ҳнҷ”н•ң кІғ

	в—Ұ мҰү, ScoreлҠ” **нҶ нҒ° к°„ кҙҖкі„(м—°кҙҖм„ұ)**лҘј лӮҳнғҖлғ…лӢҲлӢӨ.
---
### 3) мҠӨмјҖмқјл§Ғ(Scaling) : 
### $ [ \text{Scaled Score} = \frac{QK^\top}{\sqrt{d_k}} ] $
### м„ӨлӘ…)
- мҠӨмјҖмқјл§Ғ
$ [ \text{Scaled Score} = \frac{QK^\top}{\sqrt{d_k}} ] $
вҶ’ к°’мқҙ л„Ҳл¬ҙ м»Өм§Җм§Җ м•ҠлҸ„лЎқ Key м°ЁмӣҗмҲҳлЎң лӮҳлҲ  м•Ҳм •нҷ”н•©лӢҲлӢӨ.
---
### 4) Softmax ліҖнҷҳ : 
### $ [ \text{Attention Weights} = \text{Softmax}(\text{Scaled Score}) ] $
### м„ӨлӘ…)
- Softmax ліҖнҷҳ
$ [ \text{Attention Weights} = \text{Softmax}(\text{Scaled Score}) ] $ 
вҶ’ к°Ғ нҶ нҒ°мқҙ лӢӨлҘё нҶ нҒ°мқ„ м–јл§ҲлӮҳ м°ёкі н•ҙм•ј н•ҳлҠ”м§Җ нҷ•лҘ  л¶„нҸ¬лЎң ліҖнҷҳн•©лӢҲлӢӨ.
---
### 5) Value к°ҖмӨ‘н•© (мөңмў… м¶ңл Ҙ) : $ [ \text{Output} = \text{Attention Weights} \cdot V ] $
### м„ӨлӘ…)
- Attention WeightsлҘј Valueм—җ кіұн•ҙ мөңмў… м¶ңл Ҙмқ„ м–»мҠөлӢҲлӢӨ : 
$ [ \text{Output} = \text{Attention Weights} \cdot V ] $

мҰү, Query/Key/Value мғқм„ұ вҶ’ QueryмҷҖ KeyлЎң мң мӮ¬лҸ„ вҶ’ мҠӨмјҖмқјл§Ғ вҶ’ SoftmaxлЎң нҷ•лҘ нҷ” вҶ’ к·ё нҷ•лҘ лЎң ValueлҘј к°ҖмӨ‘н•©н•ҳлҠ” кІғмқҙ Self-Attentionмқҳ н•өмӢ¬мһ…лӢҲлӢӨ.

## 2. Transformer(1.Self-Attention) м–ём–ҙ лӘЁлҚёмқҳ Self-Attention кі„мӮ° кіјм •мқ„ кіөмӢқкіј н•Ёк»ҳ лӢЁкі„лі„лЎң кі„мӮ°
### 1) мһ…л Ҙ лІЎн„°, xк°’ м„Өм •
$ [ X = \begin{bmatrix}1 & 0 \\ 0 & 1\end{bmatrix} ] $

---
### 2) к°ҖмӨ‘м№ҳ н–үл ¬, мҙҲкё° к°ҖмӨ‘м№ҳ лһңлҚӨ м„Өм •
$ [ W_Q = \begin{bmatrix}1 & 0 \\ 0 & 1\end{bmatrix}, \quad
W_K = \begin{bmatrix}1 & 1 \\ 0 & 1\end{bmatrix}, \quad
W_V = \begin{bmatrix}1 & 0 \\ 0 & 1\end{bmatrix}
] $

---
### 3) Query, Key, Value кі„мӮ°(мғқм„ұ)
$ [ Q = XW_Q = \begin{bmatrix}1 & 0 \\ 0 & 1\end{bmatrix} ] $
$ [ K = XW_K = \begin{bmatrix}1 & 1 \\ 0 & 1\end{bmatrix} ] $
$ [ V = XW_V = \begin{bmatrix}1 & 0 \\ 0 & 1\end{bmatrix} ] $

---
### 4) Score(мң мӮ¬лҸ„) кі„мӮ°
$ [ QK^\top = \begin{bmatrix}1 & 0 \\ 0 & 1\end{bmatrix} \cdot \begin{bmatrix}1 & 0 \\ 1 & 1\end{bmatrix}
= \begin{bmatrix}1 & 0 \\ 1 & 1\end{bmatrix}
 ] $

---
### 5) мҠӨмјҖмқјл§Ғ
$ [ d_k = 2 \quad \Rightarrow \quad \sqrt{d_k} = 1.414 ] $
$ [ \text{Scaled Score} = \frac{1}{1.414} \cdot \begin{bmatrix}1 & 0 \\ 1 & 1\end{bmatrix}
= \begin{bmatrix}0.71 & 0 \\ 0.71 & 0.71\end{bmatrix}
 ] $

---
### 6) Softmax м Ғмҡ©
$ вҖў мІ« лІҲм§ё н–ү: [ [0.71, 0] ] вҶ’ Softmax вүҲ [ [0.67, 0.33] ] $

$ вҖў л‘җ лІҲм§ё н–ү: [ [0.71, 0.71] ] вҶ’ Softmax = [ [0.5, 0.5] ] $

л”°лқјм„ң:
$ [ \text{Attention Weights} = \begin{bmatrix}0.67 & 0.33 \\ 0.5 & 0.5\end{bmatrix} ] $

---
### 7) Value к°ҖмӨ‘н•© (мөңмў… лӢЁкі„)
$ [ \text{Output} = \text{Attention Weights} \cdot V ] $
$ [ = \begin{bmatrix}0.67 & 0.33 \\ 0.5 & 0.5\end{bmatrix} \cdot \begin{bmatrix}1 & 0 \\ 0 & 1\end{bmatrix}
= \begin{bmatrix}0.67 & 0.33 \\ 0.5 & 0.5\end{bmatrix}
 ] $

---
### 8) мөңмў… кІ°кіј
Self-Attention м¶ңл Ҙ:
$ [ \text{Output} = \begin{bmatrix}0.67 & 0.33 \\ 0.5 & 0.5\end{bmatrix} ] $

---
мқҙл ҮкІҢ н•ҙм„ң Self-AttentionмқҖ Query/Key/Value мғқм„ұ вҶ’ QKбөҖ(мң мӮ¬лҸ„ кі„мӮ°) вҶ’ мҠӨмјҖмқјл§Ғ вҶ’ Softmax вҶ’ Value к°ҖмӨ‘н•©мңјлЎң мҷ„м„ұ.

мқҙм „м—җ MLPм—җм„ң л°°мҡҙ вҖңм„ нҳ•кІ°н•© вҶ’ Softmax вҶ’ кІ°кіјвҖқ нқҗлҰ„мқҙ к·ёлҢҖлЎң нҷ•мһҘлҗң кө¬мЎ°лқјлҠ” м җмқҙ н•өмӢ¬.
