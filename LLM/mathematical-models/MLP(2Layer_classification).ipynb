# MLP(Multi-Layer Perceptron, 다층 퍼셉트론 분류 모델) 2층 분류모델 총정리

## 전체 흐름 요약
#### 1. 순전파: 입력 → 은닉층 선형결합 → ReLU → 출력층 선형결합 → Softmax → 손실 계산
#### 2. 역전파: 손실 → 출력층 오차 → 출력층 기울기 → 은닉층 오차 → 은닉층 기울기

## MLP Classification 특징
#### 1. 입력층(Input Layer): 데이터 벡터 [ x ]
#### 2. 은닉층(Hidden Layer): 선형결합 + 비선형 활성화(ReLU 등)
#### 3. 출력층(Output Layer): 선형결합 + Softmax → 확률 분포
#### 4. 손실 함수(Loss Function): Cross-Entropy → 정답 클래스 확률을 최대화하도록 학습

## 모델 정의:

### 순전파 공식
##### 1. 은닉층 선형결합 : $ [ z_1 = x W_1 + b_1 ] $
##### 2. 은닉층 활성화 (예: ReLU) : $ [ h = f(z_1) = \text{ReLU}(z_1) ] $
##### 3. 출력층 선형결합 : $ [ z_2 = h W_2 + b_2 ] #
##### 4. 소프트맥스 변환 : $ [ y_k = \frac{e^{z_{2,k}}}{\sum_j e^{z_{2,j}}} ] $
---
##### 5. 손실 함수 (Cross-Entropy)
##### 정답 클래스 [ c ]일 때 : $ [ L = -\log(y_c) ] $
---
### 역전파 공식
##### 1. 출력층 오차 : $ [ \delta_2 = y - y_{\text{true}} ] $
##### 2. 출력층 기울기 : $ [ \frac{\partial L}{\partial W_2} = h^\top \delta_2 ] $
##### 3. 은닉층 오차 : [ \delta_1 = \delta_2 W_2^\top \cdot f^\prime(z_1) ] $
##### 4. 은닉층 기울기 : $ [ \frac{\partial L}{\partial W_1} = x^\top \delta_1 ] $
---
### 경사하강법(Gradient Descent) 공식(가중치 업데이트)
#### $ [ \eta ] $ : 학습률(Learning Rate), $ [ \frac{\partial L}{\partial W} ] $ : 손실 함수에 대한 가중치의 기울기(gradient)
#### 손실 [ L ]을 최소화하기 위해 가중치 [ W ]를 업데이트하는 규칙은 다음과 같습니다 : $ [ W^{(new)} = W^{(old)} - \eta \cdot \frac{\partial L}{\partial W} ] $
#### 바이어스도 동일하게 업데이트합니다: $ [ b^{(new)} = b^{(old)} - \eta \cdot \frac{\partial L}{\partial b} ] $
---
#### MLP Classification 모델 목적 :
##### MLP Classification은 입력을 받아 은닉층을 거쳐 Softmax로 확률 분포를 만들고, 
##### Cross-Entropy 손실을 통해 정답 클래스 확률을 최대화하도록 학습합니다. 
##### 역전파는 출력층 오차에서 시작해 은닉층으로 전달되며, 각 층의 가중치 기울기를 계산해 업데이트합니다.
---

## MLP 2층 분류 모델의 순전파 → 손실 계산 → 역전파 → 경사하강법 업데이트 과정을 공식과 함께 단계별로 계산

## 1. 순전파 (Forward Propagation)
---
### 1) 입력층, 초기 설정
##### 입력 x : $ [ x = [1, 2] ] $
##### 초기 가중치 : $ [ W_1 = \begin{bmatrix}0.1 & 0.2 \\ 0.3 & 0.4\end{bmatrix} ] $, $ [ W_2 = \begin{bmatrix}0.5 & 0.6 \\ 0.7 & 0.8\end{bmatrix} ] $, 여기서 b(바이어스)는 계산을 단순화를 위해 0 또는 제외한다.
---
### 2) 은닉층
##### 은닉층 선형결합 : $ [ z_1 = x W_1 + b_1 ] $
##### $ [ z_1 = x W_1 = [1, 2] \cdot \begin{bmatrix}0.1 & 0.2 \\ 0.3 & 0.4\end{bmatrix} ] $
##### $ [ z_1 = [1\cdot0.1 + 2\cdot0.3, \; 1\cdot0.2 + 2\cdot0.4] = [0.7, 1.0] ] $
##### 은닉층 활성화 (ReLU), 비선형 함수를 적용(복잡한 패턴 학습 목적)
##### $ [ h = \text{ReLU}(z_1) = [\max(0,0.7), \max(0,1.0)] = [0.7, 1.0] ] $
---
### 3) 출력층 
#### 출력층 선형결합 : $ [ z_2 = h W_2 = [0.7, 1.0] \cdot \begin{bmatrix}0.5 & 0.6 \\ 0.7 & 0.8\end{bmatrix} ] $
#### $ [ z_2 = [0.7\cdot0.5 + 1.0\cdot0.7, \; 0.7\cdot0.6 + 1.0\cdot0.8] ] $
#### $ [ = [1.05, 1.22] ] $
---
#### 출력층 소프트맥스 변환 : $ [ y_1 = \frac{e^{1.05}}{e^{1.05}+e^{1.22}}, \quad y_2 = \frac{e^{1.22}}{e^{1.05}+e^{1.22}} ] $
#### 근사값: $ [ y_1 \approx 0.459, \quad y_2 \approx 0.544 ] $
---
### 4) 손실 함수, Loss 계산 (Cross-Entropy)
#### 정답 클래스 = 2 → $ [ L = -\log(y_2) = -\log(0.544) \approx 0.609 ] $

---
## 2. 역전파 (Backpropagation)
---
### 1) 출력층
#### 출력층 오차 계산 : 소프트맥스 + Cross-Entropy 조합의 결과 : $ [ \delta_2 = y - y_{\text{true}} = [0.459, 0.544] - [0, 1] = [0.459, -0.456] ] $
---
#### 출력층 기울기 벡터 계산 : $ [ \frac{\partial L}{\partial W_2} = h^\top \cdot \delta_2 ] $
#### $ [ = \begin{bmatrix}0.7 \\ 1.0\end{bmatrix} \cdot \begin{bmatrix}0.459 & -0.456\end{bmatrix} ] $
#### $ [ = \begin{bmatrix}0.321 & -0.319 \\ 0.459 & -0.456\end{bmatrix} ] $
#### 출력층 바이어스 기울기 벡터 계산: $ [ \frac{\partial L}{\partial b_2} = \delta_2 = [0.459, -0.456] ] $
---
### 2) 은닉층
#### 은닉층 오차 계산 : 
#### 출력층 오차를 은닉층으로 전달 : $[ \delta_1 = \delta_2 W_2^\top \cdot f^\prime(z_1) ] $
#### 여기서 $ [ f^\prime(z_1) ] $ 는 ReLU 미분 → $ [ z_1 = [0.7, 1.0] ] $ 모두 양수 → $ [ f^\prime(z_1) = [1, 1] ] $
#### $ [ \delta_1 = [0.459, -0.456] \begin{bmatrix}0.5 & 0.7 \\ 0.6 & 0.8\end{bmatrix} ] $
#### $ [ = [0.459\cdot0.5 + (-0.456)\cdot0.6, \; 0.459\cdot0.7 + (-0.456)\cdot0.8] ] $
#### $ [ = [0.2295 - 0.2736, \; 0.3213 - 0.3648] = [-0.0441, -0.0435] ] $
---
#### 은닉층 기울기 벡터 계산 :
#### $ [ \frac{\partial L}{\partial W_1} = x^\top \cdot \delta_1 ] $
#### $ [ = \begin{bmatrix}1 \\ 2\end{bmatrix} \cdot \begin{bmatrix}-0.0441 & -0.0435\end{bmatrix} ] $
#### $ [ = \begin{bmatrix}-0.0441 & -0.0435 \\ -0.0882 & -0.0870\end{bmatrix} ] $
#### 은닉층 바이어스 기울기: $ [ \frac{\partial L}{\partial b_1} = \delta_1 = [-0.0441, -0.0435] ] $
---
### 3) 역전파 요약
#### 출력층 오차: $ [ \delta_2 = [0.459, -0.456] ] $
#### 출력층 기울기 : $ [ \frac{\partial L}{\partial W_2} = \begin{bmatrix}0.321 & -0.319 \\ 0.459 & -0.456\end{bmatrix} ] $
#### 은닉층 오차 : $ [ \delta_1 = [-0.0441, -0.0435] ] $
#### 은닉층 기울기: $ [ \frac{\partial L}{\partial W_1} = \begin{bmatrix}-0.0441 & -0.0435 \\ -0.0882 & -0.0870\end{bmatrix} ] $
---
## 3. 경사하강법(Gradient Descent), 가중치 업데이트, 손실값을 줄이는 목적
---
### 1) 출력층 기울기
#### $ [ \frac{\partial L}{\partial W_2} = \begin{bmatrix}0.321 & -0.319 \\ 0.459 & -0.456\end{bmatrix} ] $
#### 학습률 $ [ \eta = 0.1 ] $ 이라고 하면 :
#### $ [ W_2^{(new)} = W_2^{(old)} - 0.1 \cdot \frac{\partial L}{\partial W_2} ] $
#### $ [ = \begin{bmatrix}0.5 & 0.6 \\ 0.7 & 0.8\end{bmatrix} - 0.1 \cdot \begin{bmatrix}0.321 & -0.319 \\ 0.459 & -0.456\end{bmatrix} ] $
#### $ [ = \begin{bmatrix}0.5 - 0.0321 & 0.6 + 0.0319 \\ 0.7 - 0.0459 & 0.8 + 0.0456\end{bmatrix} ] $
#### $ [ = \begin{bmatrix}0.468 & 0.632 \\ 0.654 & 0.846\end{bmatrix} ] $
---
### 2) 은닉층 기울기
##### $ [ \frac{\partial L}{\partial W_1} = \begin{bmatrix}-0.0441 & -0.0435 \\ -0.0882 & -0.0870\end{bmatrix} ] $
##### 학습률 $ [ \eta = 0.1 ] $ 이라고 하면 :
##### $ [ W_1^{(new)} = W_1^{(old)} - 0.1 \cdot \frac{\partial L}{\partial W_1} ] $
##### $ [ = \begin{bmatrix}0.1 & 0.2 \\ 0.3 & 0.4\end{bmatrix} - 0.1 \cdot \begin{bmatrix}-0.0441 & -0.0435 \\ -0.0882 & -0.0870\end{bmatrix} ] $
##### $ [ = \begin{bmatrix}0.1 + 0.00441 & 0.2 + 0.00435 \\ 0.3 + 0.00882 & 0.4 + 0.00870\end{bmatrix} ] $
##### $ [ = \begin{bmatrix}0.104 & 0.204 \\ 0.309 & 0.409\end{bmatrix} ] $
