# Transformer(3.Encoder Block) 언어 모델 총정리 

## 1. Encoder Block
	◦ 트랜스포머 인코더 블록 단계별 공식
	◦ Residual Connection → Layer Normalization → Feead Forward Network(FFN) → 2차 Residual Connection → 2차 Layer Normalization

---
### 1) 멀티헤드 어텐션 출력
앞서 계산한 최종 출력:
$ [ \text{MultiHead}(X) = \begin{bmatrix}1.34 & 0.66 \\ 1.0 & 1.0\end{bmatrix} ] $

---
### 2) Residual Connection (잔차 연결)
입력 $ [ X ] $ 와 멀티헤드 어텐션 출력을 더합니다.
$ [ H = X + \text{MultiHead}(X) ] $

---
### 3) Layer Normalization
LayerNorm은 각 행(토큰)에 대해 평균과 분산을 계산해 정규화합니다.

입력 벡터 $ [ x = [x_1, x_2, \ldots, x_n] ] $ 에 대해:

3-1) 평균 (Mean)
$ [ \mu = \frac{1}{n} \sum_{i=1}^{n} x_i ] $

3-2) 분산 (Variance)
$ [ \sigma^2 = \frac{1}{n} \sum_{i=1}^{n} (x_i - \mu)^2 ] $

3-3) 표준편차 (Standard Deviation)
$ [ \sigma = \sqrt{\sigma^2} ] $

3-4) 정규화 (Normalization)
$ [ \hat{x}_i = \frac{x_i - \mu}{\sigma} ] $

---
### 4) Feed Forward Network (FFN)
FFN은 두 개의 선형 변환과 비선형 활성화(ReLU)를 포함합니다.

공식: $ [ \text{FFN}(x) = \max(0, xW_1 + b_1)W_2 + b_2 ] $

4-1) 선형변환(차원 확장) : $ [ xW_1 + b_1 ] $

4-2) ReLU(비선형 함수로 복작한 패턴 학습) : $ [ \text{FFN}(x) = \max(0, xW_1 + b_1)] $

4-3) 두번째 선형변환(차원 축소) : $ [ \text{FFN}(x) = \max(0, xW_1 + b_1)W_2 + b_2 ] $

---
### 5) 트랜스포머 인코더 블록 전체 과정:
	◦ 1. Multi-Head Attention
	◦ 2. Residual Connection: 입력과 더하기
	◦ 3. LayerNorm: 평균, 분산, 표준편차, 정규화
	◦ 4. FFN: 선형변환, 비선형 변환, 두번째 선형변환

---

## 2. Transformer(3.Encoder Block) 언어 모델의 Encoder Block 계산 과정을 공식과 함께 단계별로 계산
---
### 1) 멀티헤드 어텐션 출력
앞서 계산한 최종 출력:

$ [ \text{MultiHead}(X) = \begin{bmatrix}1.34 & 0.66 \\ 1.0 & 1.0\end{bmatrix} ] $

---
### 2) Residual Connection (잔차 연결)
입력 $ [ X ] $ 와 멀티헤드 어텐션 출력을 더합니다.

$ [ H = X + \text{MultiHead}(X) ] $

계산:

$ [ X = \begin{bmatrix}1 & 0 \\ 0 & 1\end{bmatrix}, \quad \text{MultiHead}(X) = \begin{bmatrix}1.34 & 0.66 \\ 1.0 & 1.0\end{bmatrix} ] $

$ [ H = \begin{bmatrix}1+1.34 & 0+0.66 \\ 0+1.0 & 1+1.0\end{bmatrix}
= \begin{bmatrix}2.34 & 0.66 \\ 1.0 & 2.0\end{bmatrix}
 ] $

---
### 3) Layer Normalization
LayerNorm은 각 행(토큰)에 대해 평균과 분산을 계산해 정규화합니다.

예시로 첫 번째 행 $ [ [2.34, 0.66] ] $ :

3-1) 평균: $ [ (2.34+0.66)/2 = 1.5 ] $

3-2) 분산: $ [ ((2.34-1.5)^2 + (0.66-1.5)^2)/2 = (0.84^2 + (-0.84)^2)/2 = 0.7056 ] $

3-3) 표준편차: $ [ \sqrt{0.7056} \approx 0.84 ] $

3-4) 정규화:
$ [ \left[\frac{2.34-1.5}{0.84}, \frac{0.66-1.5}{0.84}\right] = [1.0, -1.0] ] $

두 번째 행 $ [ [1.0, 2.0] ] $ :

• 평균: 1.5

• 분산: 0.25

• 표준편차: 0.5

정규화:
$ [ \left[\frac{1.0-1.5}{0.5}, \frac{2.0-1.5}{0.5}\right] = [-1.0, 1.0] ] $

따라서 LayerNorm 결과:

$ [ H_{norm} = \begin{bmatrix}1.0 & -1.0 \\ -1.0 & 1.0\end{bmatrix} ] $

---
### 4) Feed Forward Network (FFN)
FFN은 두 개의 선형 변환과 비선형 활성화(ReLU)를 포함합니다.
공식:

$ [ \text{FFN}(x) = \max(0, xW_1 + b_1)W_2 + b_2 ] $

4-1) 선형변환(차원 확장) : $ [ xW_1 + b_1 ] $

4-2) ReLU(비선형 함수로 복작한 패턴 학습) : $ [ \text{FFN}(x) = \max(0, xW_1 + b_1)] $

4-3) 두번째 선형변환(차원 축소) : $ [ \text{FFN}(x) = \max(0, xW_1 + b_1)W_2 + b_2 ] $

---
예시 1):
앞서 LayerNorm 결과:
$ [ H_{norm} = \begin{bmatrix}1.0 & -1.0 \\ -1.0 & 1.0\end{bmatrix} ] $

예시 가중치:
$ [ W_1 = \begin{bmatrix}1 & 0 \\ 0 & 1\end{bmatrix}, \quad b_1 = 0 ] $
$ [ W_2 = \begin{bmatrix}1 & 0 \\ 0 & 1\end{bmatrix}, \quad b_2 = 0 ] $

1. 첫 번째 선형 변환
$ [ H_{norm} W_1 + b_1 = H_{norm} = \begin{bmatrix}1.0 & -1.0 \\ -1.0 & 1.0\end{bmatrix} ] $

2. ReLU 적용
$ [ \max(0, H_{norm}) = \begin{bmatrix}1.0 & 0.0 \\ 0.0 & 1.0\end{bmatrix} ] $

3. 두 번째 선형 변환
$ [ \begin{bmatrix}1.0 & 0.0 \\ 0.0 & 1.0\end{bmatrix} W_2 = \begin{bmatrix}1.0 & 0.0 \\ 0.0 & 1.0\end{bmatrix} ] $

4. 최종 FFN 출력
$ [ \text{FFN}(H_{norm}) = \begin{bmatrix}1.0 & 0.0 \\ 0.0 & 1.0\end{bmatrix} ] $

---
예시 2):
앞서 LayerNorm 결과:
$ [ H_{norm} = \begin{bmatrix}1.0 & -1.0 \\ -1.0 & 1.0\end{bmatrix} ] $

예시로 $ [ W_1, W_2 ] $ 를 랜덤 초기값 대신 단순한 행렬로 설정:

$ [ W_1 = \begin{bmatrix}2 & -1 \\ 1 & 3\end{bmatrix}, \quad b_1 = \begin{bmatrix}0 & 0\end{bmatrix} ] $

$ [ W_2 = \begin{bmatrix}1 & 0 \\ 0 & 1\end{bmatrix}, \quad b_2 = \begin{bmatrix}0 & 0\end{bmatrix} ] $

1. 첫 번째 선형 변환
$ [ H_{norm} W_1 = \begin{bmatrix}1.0 & -1.0 \\ -1.0 & 1.0\end{bmatrix} \cdot \begin{bmatrix}2 & -1 \\ 1 & 3\end{bmatrix} ] $

계산:

• 첫 번째 행: $ [ [1,-1] \cdot W_1 = [2-1, -1-3] = [1, -4] ] $

• 두 번째 행: $ [ [-1,1] \cdot W_1 = [-2+1, 1* -1 + 3] = [-1, 2] ] $

$ [ = \begin{bmatrix}1 & -4 \\ -1 & 2\end{bmatrix} ] $

2. ReLU 적용
$ [ \max(0, \begin{bmatrix}1 & -4 \\ -1 & 2\end{bmatrix}) = \begin{bmatrix}1 & 0 \\ 0 & 2\end{bmatrix} ] $

3. 두 번째 선형 변환
$ [ \begin{bmatrix}1 & 0 \\ 0 & 2\end{bmatrix} W_2 = \begin{bmatrix}1 & 0 \\ 0 & 2\end{bmatrix} ] $

4. 최종 FFN 출력
$ [ \text{FFN}(H_{norm}) = \begin{bmatrix}1 & 0 \\ 0 & 2\end{bmatrix} ] $

	◦ [ W_1, W_2 ]는 초기에 랜덤 값으로 설정되는 학습 파라미터

	◦ 학습 과정에서 손실 함수를 최소화하도록 업데이트됨

	◦ FFN은 입력을 확장 → ReLU → 축소하는 구조로, 모델이 더 복잡한 패턴을 학습할 수 있게 해줌
