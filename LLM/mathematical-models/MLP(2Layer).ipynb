# MLP(Multi-Layer Perceptron) 2층 모델 총정리

## 모델 정의:

• 은닉층: $ [  h = f(XW_1 + b_1)  ] $

• 출력층: $ [  y = hW_2 + b_2  ] $

• 손실 함수(MSE):
$ [ L = \frac{1}{2}(y_{\text{true}} - y)^2 ] $

---
## 순전파(Forward Pass):

1. 은닉층 출력 계산:
$ [ h = f(XW_1 + b_1) ] $
2. 출력층 출력 계산:
$ [ y = hW_2 + b_2 ] $
3. 손실 함수 계산:
$ [ L = \frac{1}{2}(y_{\text{true}} - y)^2 ] $
---
## 역전파(Backward Pass):

1. 출력층 오차:
$ [ \frac{\partial L}{\partial y} = (y - y_{\text{true}}) ] $
2. 출력층 기울기:
$ [ \frac{\partial L}{\partial W_2} = (y - y_{\text{true}}) \cdot h ] $
$ [ \frac{\partial L}{\partial b_2} = (y - y_{\text{true}}) ] $
3. 은닉층 오차:
$ [ \frac{\partial L}{\partial h} = (y - y_{\text{true}}) \cdot W_2 ] $
4. 은닉층 기울기:
$ [ \frac{\partial L}{\partial W_1} = X \cdot \frac{\partial L}{\partial h} ] $
$ [ \frac{\partial L}{\partial b_1} = \frac{\partial L}{\partial h} ] $
---
## 경사하강법(Gradient Descent):
손실을 줄이기 위해 각 파라미터를 기울기의 반대 방향으로 업데이트합니다. 학습률 [ \eta ]를 곱해 이동 크기를 조절합니다.
업데이트 공식:
1. 출력층 기울기:
$ [ W_2 \leftarrow W_2 - \eta \frac{\partial L}{\partial W_2} ] $
$ [ b_2 \leftarrow b_2 - \eta \frac{\partial L}{\partial b_2} ] $
2. 은닉층 기울기:
$ [ W_1 \leftarrow W_1 - \eta \frac{\partial L}{\partial W_1} ] $
$ [ b_1 \leftarrow b_1 - \eta \frac{\partial L}{\partial b_1} ] $
---
즉, MLP 2층 모델은 순전파 → 손실 계산 → 역전파 → 경사하강법 업데이트 과정을 반복하면서 최적의 [ W_1, b_1, W_2, b_2 ]를 찾아가는 구조입니다.

### 예시 데이터를 가지고 MLP 2층 모델의 순전파 → 손실 계산 → 역전파 → 경사하강법 업데이트 과정을 공식과 함께 단계별로 계산
---
예시 조건:
• 입력 $ [ x = 2 ] $

• 목표값 $ [ y_{\text{true}} = 10 ] $

• 초기 파라미터: $ [ W_1 = 3, b_1 = 1, W_2 = 4, b_2 = 1 ] $

• 학습률 $ [ \eta = 0.1 ] $

---
(1) 순전파 (Forward Propagation)

은닉층 출력: 

$ [ h = xW_1 + b_1 = 2 \cdot 3 + 1 = 7 ] $

출력층 출력: 

$ [ y = hW_2 + b_2 = 7 \cdot 4 + 1 = 29 ] $

손실 함수: 

$ [ L = \frac{1}{2}(y_{\text{true}} - y)^2 = \frac{1}{2}(10 - 29)^2 = \frac{1}{2} \cdot 361 = 180.5 ] $

---
(2) 역전파 (Backward Propagation)

출력층 오차: 

$ [ \frac{\partial L}{\partial y} = (y - y_{\text{true}}) = (29 - 10) = 19 ] $

출력층 기울기: 

$ [ \frac{\partial L}{\partial W_2} = (y - y_{\text{true}}) \cdot h = 19 \cdot 7 = 133 ] $

$ [ \frac{\partial L}{\partial b_2} = (y - y_{\text{true}}) = 19 ] $

은닉층 오차:

$ [ \frac{\partial L}{\partial h} = (y - y_{\text{true}}) \cdot W_2 = 19 \cdot 4 = 76 ] $

은닉층 기울기:

$ [ \frac{\partial L}{\partial W_1} = x \cdot \frac{\partial L}{\partial h} = 2 \cdot 76 = 152 ] $

$ [ \frac{\partial L}{\partial b_1} = \frac{\partial L}{\partial h} = 76 ] $

---
(3) 경사하강법 업데이트 (Gradient Descent Update)
업데이트 공식:
출력층 기울기 업데이트:

$ [ W_2 \leftarrow W_2 - \eta \frac{\partial L}{\partial W_2} = 4 - 0.1 \cdot 133 = -9.3 ] $

$ [ b_2 \leftarrow b_2 - \eta \frac{\partial L}{\partial b_2} = 1 - 0.1 \cdot 19 = -0.9 ] $

은닉층 기울기 업데이트:

$ [ W_1 \leftarrow W_1 - \eta \frac{\partial L}{\partial W_1} = 3 - 0.1 \cdot 152 = -12.2 ] $

$ [ b_1 \leftarrow b_1 - \eta \frac{\partial L}{\partial b_1} = 1 - 0.1 \cdot 76 = -6.6 ] $

---
(4) 결과
• 업데이트 후 파라미터:

	◦ [ W_1 = -12.2 ]
	◦ [ b_1 = -6.6 ]
	◦ [ W_2 = -9.3 ]
	◦ [ b_2 = -0.9 ]
---
### 이렇게 입력 [ x=2 ], 목표값 [ y_{\text{true}}=10 ]으로 시작한 MLP 2층 모델은 순전파 → 손실 계산 → 역전파 → 경사하강법 업데이트 과정을 거쳐 새로운 파라미터 값으로 갱신됩니다.
